## Retour sur la conversion des titres et des tickets

```python
df3
```

```python
import re

def convert_ticket(raw_ticket):
    try:
        res = re.findall("\d+$", raw_ticket)
        return res[0]
    except:
        print(raw_ticket)
        return -1
```

```python
df3.Ticket = df3.Ticket.map(convert_ticket)
```

```python
df3 = df3[df3.Ticket != -1]
```

```python
df3.count()
```

```python
import re

def convert_title(raw_title):
    try:
        res = re.findall(", ([^.]*)\.", raw_title)
        return res[0]
    except:
        print(raw_ticket)
        return -1
```

```python
df3["Title"] = df3.Name.map(convert_title)
```
